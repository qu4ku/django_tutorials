```python
df = pd.DataFrame(np.random.randn(600, 3), index=pd.date_range('5/1/2016', 
    freq='D', periods=600), columns=['A', 'B', 'C'])
df.tail()
```

```python
r = df.rolling(window=10)
df.plot(style='k--')
r.mean().plot(style='k')
```

```python
df['A'].plot(color='grey')
r.quantile(.30)['A'].plot(color='red')
```
